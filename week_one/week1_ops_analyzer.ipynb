## Question 1 – KPI Function: `calculate_efficiency`

**What this function does**

Efficiency tells us *how close* a site came to hitting its target output.
The formula is straightforward:

$$\text{Efficiency (\%)} = \frac{\text{Actual Output}}{\text{Target Output}} \times 100$$

**Parameters**
- `actual_output` – the units actually produced / delivered during the period.
- `target_output` – the planned / expected units for the same period.

**Returns** a `float` representing the efficiency percentage (e.g. `72.0`).

> 💡 *Junior Analyst Tip:* We multiply by 100 so the result is already in
> percentage form — no need to multiply again when you print it.


```
def calculate_efficiency(actual_output, target_output):
    """Return the efficiency percentage of a site."""
    efficiency = (actual_output / target_output) * 100
    return efficiency
```

---
## Question 2 – Status Classifier: `get_operational_status`

Once we have the efficiency percentage we need to translate it into an
actionable status label so operations teams know *at a glance* where to focus.

| Efficiency Range | Status |
|---|---|
| Below 70 % | 🔴 **Critical** |
| 70 % – 89 % | 🟡 **Warning** |
| 90 % and above | 🟢 **Normal** |

We use `if / elif / else` to check the thresholds in order from most severe to
least severe.


```
def get_operational_status(efficiency):
    """Classify operational status based on efficiency percentage."""
    if efficiency < 70:
        return "Critical"
    elif efficiency < 90:       # covers 70 – 89.99...
        return "Warning"
    else:
        return "Normal"
```

---
## Question 3 – Data Modelling: Site Dictionaries

We store each depot as a **dictionary** with three keys:
- `name` – a human‑readable label for the site.
- `actual_output` – litres of fuel dispatched this period.
- `target_output` – planned litres for the same period.

All three dictionaries are collected in a **list** called `sites` so we can
loop over them easily in the next step.


```
sites = [
    {
        "name": "Nairobi Fuel Depot",
        "actual_output": 7200,
        "target_output": 10000
    },
    {
        "name": "Mombasa Fuel Depot",
        "actual_output": 9100,
        "target_output": 10000
    },
    {
        "name": "Kisumu Fuel Depot",
        "actual_output": 6200,
        "target_output": 10000
    }
]
```

---
## Question 4 – Execution: Loop & Report

Now we bring everything together.

**What the loop does — step by step**
1. **Iterates** over every site dictionary in `sites`.
2. **Unpacks** `actual_output` and `target_output` from the dictionary.
3. **Calls** `calculate_efficiency()` to get the percentage.
4. **Calls** `get_operational_status()` to get the status label.
5. **Prints** a tidy, formatted report for the site.

> 💡 *Junior Analyst Tip:* `:.2f` inside an f‑string formats a float to
> exactly two decimal places — useful for clean report output.


```
print("=" * 45)
print("     FUEL DEPOT OPERATIONS REPORT")
print("=" * 45)

for site in sites:
    name          = site["name"]
    actual_output = site["actual_output"]
    target_output = site["target_output"]

    efficiency = calculate_efficiency(actual_output, target_output)
    status     = get_operational_status(efficiency)

    print(f"\nSite         : {name}")
    print(f"Actual Output: {actual_output:,} litres")
    print(f"Target Output: {target_output:,} litres")
    print(f"Efficiency   : {efficiency:.2f}%")
    print(f"Status       : {status}")
    print("-" * 45)

print("\nReport complete.")
```

## Question 5 – Summary & Documentation

### What we built

| Component | Purpose |
|---|---|
| `calculate_efficiency()` | Converts raw output figures into a % KPI |
| `get_operational_status()` | Maps the KPI to a traffic‑light status |
| `sites` list of dicts | Lightweight data model — no database needed |
| `for` loop | Automates the report across any number of sites |

### Key Takeaways for Junior Analysts
- **Functions** keep logic reusable — change the threshold once, it updates everywhere.
- **Dictionaries** are great for structured records with named fields.
- **Lists of dictionaries** let you scale from 3 sites to 300 with zero code changes.
- Always format numeric output (`:,` for thousands separator, `:.2f` for decimals)
  to make reports readable at a glance.
